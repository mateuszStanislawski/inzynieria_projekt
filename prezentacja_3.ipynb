```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.stats import norm
import random
```

```python
plt.style.use('ggplot')
```

```python
euribor = pd.read_csv("euoeur3m_d.csv")
```

```python
euribor.head()
```

```python
T = 3
mu = euribor.Zamkniecie.mean()
sigma = euribor.Zamkniecie.std()
n = 250
K = n * T
dt = 1/n
```

```python
df = pd.DataFrame()
for i in range(100):
    brownian_motion = []
    W = 0
    for j in range(K):
        W = W + math.sqrt(dt) + norm.ppf(random.uniform(0, 1))
        brownian_motion.append(W)
    df[f"symulacja{i+1}"] = brownian_motion
```
